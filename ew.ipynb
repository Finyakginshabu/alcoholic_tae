```python
selected_columns = ['Q1', 'Q2', 'Q3', 'Q9', 'Q14', 'Q15', 'Q16', 'Q17', 'Q18', 'Q20', 
                    'Q23', 'Q24', 'Q25', 'Q26', 'Q27', 'Q41', 'Q42', 'Q43', 'Q44', 'Q76', 
                    'Q80', 'Q84', 'Q85', 'Q86', 'Q87', 'Q89', 'Q90', 'Q91', 'Q100', 'Q101', 
                    'Q102', 'Q103', 'Q104', 'Q105', 'Q106']
data = raw_data[selected_columns]
data.info()
```

```python
data.columns = ['age', 'gender', 'education_level', 'ride_drinking_driver', 
                'school_safety_concern', 'threatened_at_school', 'fight_count', 
                'school_fight_count', 'neighborhood_violence', 'sexual_violence', 
                'school_racism', 'school_bullying', 'cyber_bullying', 'feel_hopeless', 
                'considered_suicide', 'first_alcohol', 'alcohol_frequency', 
                'hardcore_alcohol_frequency', 'drinking_in_a_row', 'exercise', 
                'social_media', 'mental_health', 'hours_of_sleep', 'sleep_place', 
                'grade', 'parent_emotional_abuse', 'parent_physical_abuse', 
                'parent_abuse_parent', 'parent_uses_alcohol', 'parent_mental_illness', 
                'incarcerated_parent', 'have_friend', 'parent_monitoring', 
                'unfair_discipline_school', 'ADHD']
```

```python
# Mapping Values
drink_in_a_row = {
    'Did not drink in past 30 days': 1,
    '1 or 2 drinks': 2,
    '3 drinks': 3,
    '4 drinks': 4,
    '5 drinks': 5,
    '6 or 7 drinks': 6,
    '8 or 9 drinks': 7,
    '10 or more drinks': 8
}

exercise = {
    '0 days': 1,
    '1 day': 2, 
    '2 days': 3,
    '3 days': 4,
    '4 days': 5,
    '5 days': 6,
    '6 days': 7,
    '7 days': 8
}

social_media = {
    "I do not use social media": 1,
    "A few times a month": 2,
    "About once a week": 3,
    "A few times a week": 4,
    "About once a day": 5,
    "Several times a day": 6,
    "About once an hour": 7,
    "More than once an hour": 8
}

sleep_hours = {
    '4 or less hours': 1,
    '5 hours': 2,
    '6 hours': 3,
    '7 hours': 4,
    '8 hours': 5,
    '9 hours': 6,
    '10 or more hours': 7
}

housing = {
    'Home of parent or gaurdian': 1,
    'Home of other people': 2,
    'Shelter or emergency housing': 3,
    'Motel or hotel': 4,
    'Public place': 5,
    'No usual place': 6,
    'Somewhere else': 7
}

grades = {
    'Mostly A\'s': 1,
    'Mostly B\'s': 2,
    'Mostly C\'s': 3,
    'Mostly D\'s': 4,
    'Mostly F\'s': 5,
    'None of these grades': 6,
    'Not sure': 7
}

feel_close = {
    'Strongly agree': 1,
    'Agree': 2,
    'Not sure': 3,
    'Disagree': 4,
    'Strongly disagree': 5
}

frequency = {
    'Never': 1,
    'Rarely': 2,
    'Sometimes': 3,
    'Most of the time': 4,
    'Always': 5
}

yes_no = {
    'Yes': 1,
    'No': 2
}

data['drinking_in_a_row'] = data['drinking_in_a_row'].str.strip().map(drink_in_a_row)
data['exercise'] = data['exercise'].str.strip().map(exercise)
data['social_media'] = data['social_media'].str.strip().map(social_media)
data['mental_health'] = data['mental_health'].str.strip().map(frequency)
data['hours_of_sleep'] = data['hours_of_sleep'].str.strip().map(sleep_hours)
data['sleep_place'] = data['sleep_place'].str.strip().map(housing)
data['grade'] = data['grade'].str.strip().map(grades)
data['parent_emotional_abuse'] = data['parent_emotional_abuse'].str.strip().map(frequency)
data['parent_physical_abuse'] = data['parent_physical_abuse'].str.strip().map(frequency)
data['parent_abuse_parent'] = data['parent_abuse_parent'].str.strip().map(frequency)
data['parent_uses_alcohol'] = data['parent_uses_alcohol'].str.strip().map(yes_no)
data['parent_mental_illness'] = data['parent_mental_illness'].str.strip().map(yes_no)
data['incarcerated_parent'] = data['incarcerated_parent'].str.strip().map(yes_no)
data['have_friend'] = data['have_friend'].str.strip().map(feel_close)
data['parent_monitoring'] = data['parent_monitoring'].str.strip().map(frequency)
data['unfair_discipline_school'] = data['unfair_discipline_school'].str.strip().map(yes_no)
data['ADHD'] = data['ADHD'].str.strip().map(yes_no)
```

```python
data['drinking_in_a_row'] = data['drinking_in_a_row'].fillna(data['drinking_in_a_row'].median())
data['exercise'] = data['exercise'].fillna(data['exercise'].median())
data['social_media'] = data['social_media'].fillna(data['social_media'].median())
data['mental_health'] = data['mental_health'].fillna(data['mental_health'].median())
data['hours_of_sleep'] = data['hours_of_sleep'].fillna(data['hours_of_sleep'].median())
data['sleep_place'] = data['sleep_place'].fillna(0) # 0 -> UNKNOWN
# keep only who has a grade info
data = data[~data['grade'].isin([6, 7])] # drop none grade and not sure
data = data.dropna(subset=['grade']) # drop students who do not have a grade info
data['parent_emotional_abuse'] = data['parent_emotional_abuse'].fillna(data['parent_emotional_abuse'].median())
data['parent_physical_abuse'] = data['parent_physical_abuse'].fillna(data['parent_physical_abuse'].median())
data['parent_abuse_parent'] = data['parent_abuse_parent'].fillna(data['parent_abuse_parent'].median())
data['parent_uses_alcohol'] = data['parent_uses_alcohol'].fillna(data['parent_uses_alcohol'].median())
data['parent_mental_illness'] = data['parent_mental_illness'].fillna(data['parent_mental_illness'].median())
data['incarcerated_parent'] = data['incarcerated_parent'].fillna(data['incarcerated_parent'].median())
data['have_friend'] = data['have_friend'].fillna(data['have_friend'].median())
data['parent_monitoring'] = data['parent_monitoring'].fillna(data['parent_monitoring'].median())
data['unfair_discipline_school'] = data['unfair_discipline_school'].fillna(data['unfair_discipline_school'].median())
data['ADHD'] = data['ADHD'].fillna(data['ADHD'].median())
```
